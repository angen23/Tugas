```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
ubhara_link = "https://www.worldnovel.online/"
html = urlopen(ubhara_link).read()
soup = BeautifulSoup(html, 'html.parser')
categories_table = soup.find("div", {"class": "section-body"})
for q in categories_table.findAll('ul'):
    print(q.text)
```
